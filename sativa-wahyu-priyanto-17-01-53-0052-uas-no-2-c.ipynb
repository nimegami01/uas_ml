```python
# Import library pandas, digunakan untuk proses analisis data seperti manipulasi data, persiapan data, dan pembersihan data.
import pandas as pd
# Import library numpy, digunakan untuk memanajemen array, dengan konsumsi memori yang lebih kecil.
import numpy as np
#Mengimport library seaborn. Seaborn akan mempermudah kita sebagai analis data untuk memproduksi visualisasi yang indah tanpa kostumisasi rumit seperti yang kita hadapi pada matplotlib.
import seaborn as sns

# Mengimport library matplotlib. Library ini digunakan untuk visualisasi data.
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Class utama Python yang digunakan untuk mengelola data yang di olah oleh Pyhton
from IPython.core.interactiveshell import InteractiveShell
# Keluaran dari ekspresi dijadikan menjadi all, artinya semua.
InteractiveShell.ast_node_interactivity = "all"

# Import dari Kmeans, algoritma yang digunakan untuk klastering.
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# Mengambil data proyeksi penduduk yang nantinya akan di klastering.
data = pd.read_excel("../input/data-proyeksi-penduduk/proyeksi-data-penduduk-uas-no-2-c.xlsx")

# Menampilkan data yang telah di ambil, dengan batas 5 baris data.
print(f"head: \n{data.head(5)}\n")
print(f"Tail: \n {data.tail()} \n")
print(f"Describe data: \n{data.describe()}\n")
print(f"Info: \n{data.info()}\n")
```

```python
# Mengambil data dari kolom 1 sampai 7.
data_x = data.iloc[:, 1:7]
data_x.head()
```

```python
#  Melakukan proses untuk mendapatkan nilai skor (score) yang nantinya untuk klastering.
scr=[]

for i in range (1, 35):
    score = KMeans(n_clusters=i).fit(data_x).score(data_x)
    print(score)
    scr.append(score)
```

```python
# membuat garis dari proses di atas.

plt.plot(scr)
```

```python
# Melakukan klastering

kmeans = KMeans(n_clusters=5)
kmeans.fit(data_x)
```

```python
# Menambahkan nilai yang telah di proses di atas ke kolom baru, dengan label "Cluster".

data_x["Cluster"] = kmeans.labels_
data_x
```

```python
# Melakukan visualisasi klastering
plt.hist(data_x["Cluster"])
```

```python
# Melakukan visualisasi klastering ke seaborn.
sns.pairplot(data_x, hue="Cluster")
```
